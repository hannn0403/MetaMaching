```python
import pandas as pd

def extract_datafield_version(column_name):
    # column 이름에서 datafield와 version을 분리하여 반환하는 함수
    datafield, version = column_name.split('-')
    return datafield, float(version)

def keep_latest_version_per_sample(row):
    # 각 샘플(row)에서 null 값이 아닌 값들 중에서 가장 최신 버전만 남기는 함수
    latest_values = {}
    for column, value in row.items():
        if pd.notnull(value):
            datafield, version = extract_datafield_version(column)
            if datafield not in latest_values or version > latest_values[datafield][1]:
                latest_values[datafield] = (value, version)

    latest_values_series = pd.Series([latest_values[datafield][0] for datafield in latest_values],
                                     index=[datafield for datafield in latest_values])
    return latest_values_series

# 예시 DataFrame 생성 (null 값 추가)
data = {
    '23098-0.0': [1, None, 3],
    '23098-1.0': [4, 5, None],
    '23098-2.0': [None, 8, 9],
    '23099-0.0': [10, 11, None],
    '23099-1.0': [None, 14, 15],
    '23099-2.0': [16, None, 18]
}

df = pd.DataFrame(data)

# 각 샘플마다 최신 버전만 남기는 함수 호출
result_df = df.apply(keep_latest_version_per_sample, axis=1)

print(result_df)
```

```python
import os
import pandas as pd
from sklearn.decomposition import PCA

def perform_pca(file_path, n_components):
    # Read the CSV file and set the 'eid' column as the index
    df = pd.read_csv(file_path, index_col=0)
    
    
    if 'eid' in df.columns.tolist():
        df = df.set_index('eid')
        
    numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
    df = df[numeric_columns]
    print(df.shape)

    # Check if the number of columns is greater than 6
    # if df.shape[1] > 6:
        # n_components = min(6, n_components)
    # else:
        # n_components = min(df.shape[1], n_components)
    if df.shape[1] < 6:
        n_components = df.shape[1]
    else:
        n_components = 6

    # Create the PCA model and fit it on a fixed set of 1000 samples
    pca = PCA(n_components=n_components, svd_solver='auto')
    fixed_samples = df.sample(n=1000, random_state=42)
    pca.fit(fixed_samples)
    fixed_samples.to_csv(f"./ukb/imputed_files_and_pca_files_after_rearranging/sample_list_check/{file_path.split('/')[-1][:-4]}_pca_sample_list.csv")

    # Perform PCA transformation on the entire dataset
    pca_result = pca.transform(df)

    # Create a DataFrame with the PCA results, using the 'eid' index
    pca_df = pd.DataFrame(data=pca_result, index=df.index, columns=[f'PC{i+1}' for i in range(n_components)])

    return pca_df

def process_csv_files(directory_path, n_components=6):
    # Get a list of CSV files in the specified directory
    csv_files = [file for file in os.listdir(directory_path) if file.endswith('.csv')]

    # Perform PCA for each CSV file and combine the results
    # result_df = pd.concat([perform_pca(os.path.join(directory_path, file), n_components) for file in csv_files])
    for csv in csv_files: 
        result_df = perform_pca('./ukb/imputed_files_and_pca_files_after_rearranging/'+ csv, n_components)
        result_df.to_csv('./ukb/imputed_files_and_pca_files_after_rearranging/pca/'+ csv[:-4] + '_pca.csv')
        print(f"{csv} file Complete!")
    
    # result_df = pd.concat([perform_pca(os.path.join(directory_path, file), n_components) for file in csv_files])

    return result_df

# Specify the directory path containing the CSV files
csv_directory = r'C:\Users\HYUK\meta_matching\ukb\imputed_files_and_pca_files_after_rearranging'

# Perform PCA and get the final DataFrame
pca_result_df = process_csv_files(csv_directory)

# print(pca_result_df)
```
