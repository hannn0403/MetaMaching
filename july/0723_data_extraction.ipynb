### Drop Null Data 

위에서 추출한 csv 파일들에서 해당 샘플의 모든 값이 null인 샘플은 삭제한다. 

```python
def drop_null(file_path, csv_file_list, dest_path) : 
    for i in range(len(csv_file_list)):
        print(f"Dropping Null data in {csv_file_list[i]}")
        origin_df = pd.read_csv(file_path + csv_file_list[i], index_col=0)
        drop_null_df = origin_df.dropna(how='all')
        
        drop_null_df.to_csv(dest_path + csv_file_list[i])
        
    return 0
        
    
```

```python
file_path = './ukb/brain_description_sub_cat/img_feat_csv/'
csv_file_list = os.listdir(file_path)
dest_path = './ukb/brain_description_sub_cat/drop_null/'

drop_null(file_path, csv_file_list, dest_path)
```

```python
def show_file_shape(file_path):
    file_list = os.listdir(file_path)
    for file in file_list:
        df = pd.read_csv(file_path + file, index_col=0)
        print(f"{file} shape : {df.shape}")
        
    return 0 


show_file_shape('./ukb/brain_description_sub_cat/drop_null/')
```

```python
def is_the_same(file_path, file_list):
    standard_df = pd.read_csv(file_path + file_list[0], index_col=0)
    standard_idx = standard_df.index.tolist()
    
    for i in range(1, len(file_list)):
        df = pd.read_csv(file_path + file_list[i], index_col=0)
        df_idx = df.index.tolist()
        
        if standard_idx != df_idx:
            print(f"{file_list[0]} and {file_list[i]} have difference index list!")
            
    return 0

is_the_same(file_path, fs_file_list)
```

이를 통해서 FreeSurfer로 시작하는 것은 모두 동일한 subject를 가지고 있다는 것을 알 수 있다. 

```python
def df_include(file_path, big_file, small_file):
    big_df = pd.read_csv(file_path + big_file, index_col=0)
    small_df = pd.read_csv(file_path + small_file, index_col=0)
    
    big_idx = big_df.index.tolist()
    small_idx = small_df.index.tolist()
    
    big_len = len(big_idx)
    small_len = len(small_idx)
    
    big_idx_set = set(big_idx)
    small_idx_set = set(small_idx)
    
    common_elem = big_idx_set.intersection(small_idx_set)
    
    if len(common_elem) == small_len: 
        print(f"{big_file} include {small_file}")
        print(f"둘이 겹치는 샘플의 개수 : {len(common_elem)}")
    else:
        print(f"{big_file} do not includde {small_file}")
        print(f"둘이 겹치는 샘플의 개수 : {len(common_elem)}")
        
    return list(common_elem)
```
